```python
import pandas as pd

data = pd.read_csv("wineQualityReds.csv")
data.head()
```

## Line Chart

```python
# Line Chart of pH values
lc = pd.Series(data['pH'])
lc.plot()
```

## Bar Plot

```python
# Bar Chart of quality count
bp = data[["quality", "pH"]].groupby("quality").agg('count')
bp.reset_index(inplace=True)
bp = pd.Series(bp["pH"].tolist(), index = bp['quality'].tolist())
bp.plot.bar(color='k', alpha=0.7)
```

## Histogram

```python
# Histogram of pH
data['pH'].plot.density()
```

## Scatter plot

```python
# Scatter plot of 'free.sulfur.dioxide' vs 'total.sulfur.dioxide'
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure()
sns.regplot('free.sulfur.dioxide', 'total.sulfur.dioxide', data = data) 
plt.title('Changes in log %s versus log %s' % ('free.sulfur.dioxide', 'total.sulfur.dioxide'))
```

## Box Plot

```python
# Box plot of pH by wine quality
sns.catplot(x = 'quality', y = 'pH', kind = 'box', data = data)
```
