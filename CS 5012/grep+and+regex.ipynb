Helpful metacharacters [link](https://docs.python.org/3/library/re.html)
* `\b` matches word boundary
* `\d` any digit, i.e. `[0-9]`
* `\D` Any non-digit, i.e. `[^0-9]`
* `\s` any whitespace char, i.e. `[\t\n\r\f\v]`.
* `\S` any non whitespace, i.e. `[^\t\n\r\f\v]`.
* `\w` any alphanumeric char, i.e. `[a-zA-Z0-9]`.
* `\W` any non-alphanumeric char, i.e. `[^a-zA-Z0-9]`.

Helpful metacharacters, repetitions
* `*` 0 or more
* `+` 1 or more
* `?` 0 or 1
* `{n}` exactly n times
* `{n,}` at least n
* `{,n}` at most n
* `{m,n}` at least n, at most m

```python
import re

# check if an social security number is valid 
regex = r'a(ab)*a'

# Only the last test case is valid
test_cases = '''abababa
aaba
aabbaa
aba
aabababa'''

for ssn in test_cases.split('\n'):
    if(re.match(regex, ssn)):  
        print(f"{ssn} gooood  ")  
    else:
        print(f"{ssn} baaaad") 
```

```python
# check if an social security number is valid 
regex = r'ab+c?'

# Only the last test case is valid
test_cases = '''abc
ac
abbb
bbc'''

for ssn in test_cases.split('\n'):
    if(re.search(regex, ssn)):  
        print(f"{ssn} gooood  ")  
    else:
        print(f"{ssn} baaaad") 
```

```python
# check if an social security number is valid 
regex = r'[a-zA-Z]*[^,]='

# Only the last test case is valid
test_cases = '''battle!
Hot
green
swamping.
jump up.
undulate?
is.?'''

for ssn in test_cases.split('\n'):
    if(re.match(regex, ssn)):  
        print(f"{ssn} gooood")  
    else:
        print(f"{ssn} baaaad") 
```
