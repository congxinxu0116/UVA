## Logistic Regression Example: Income

Based on page provided by Peadar Coyle and J. Benjamin Cook:
https://docs.pymc.io/notebooks/GLM-logistic.html

Question: How likely is someone to earn more than $50k in income?

Data is cleaned subset of https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data 

```python
import numpy as np
import pandas as pd
import numpy.random as rnd
import seaborn as sns
from matplotlib import animation
import pymc3 as pm
import arviz as az
import matplotlib.pyplot as plt

```

```python
# Read the data

path = "/Users/donaldbrown/Dropbox/department/Classes/DS6014/CourseraBayesianML/Week4MCMC/"
file = "adult_us_postprocessed.csv"

data = pd.read_csv(path+file)
data.describe()
```

```python
data.head()
```

```python
numeric_predictors = ["age","educ",'hours']
g = sns.pairplot(data[numeric_predictors])
```

```python
# Use pm.find_MAP() to find the maximum a-posteriori (MAP) estimate for the vector of regression coefficients .

with pm.Model() as manual_logistic_model:
    Y = data["income_more_50K"]
    # intercept
    α = pm.Normal('α', mu=0, sd=100) 
    
    # coefficients
    β1 = pm.Normal('β1', mu=0, sd=100) 
    β2 = pm.Normal('β2', mu=0, sd=100)  
    
    # Equation
    μ = α + pm.math.dot(data["age"], β1)+pm.math.dot(data["educ"],β2) 
    
    θ = pm.Deterministic('θ',  pm.invlogit(μ))
    
    Yl = pm.Bernoulli('Yl', p=θ, observed=Y) 
    map_estimate = pm.find_MAP()
    print(map_estimate)
```

```python
# The estimate of the posterior distribution

with pm.Model() as logistic_model:
        data['age'] = data['age']/10.
        data['age2'] = np.square(data['age'])
        Y = data["income_more_50K"]
        pm.glm.GLM.from_formula('Y ~ sex + age + age2 + educ + hours', data, family=pm.glm.families.Binomial())
        trace = pm.sample(1000, tune=1000, init='adapt_diag', cores =4)
```

```python
with pm.Model() as logistic_model:
    az.plot_trace(trace)
```

```python
plt.figure(figsize=(9,7))
sns.jointplot(trace['age'], trace['educ'], kind="hex", color="#4CB391")
plt.xlabel("beta_age")
plt.ylabel("beta_educ");
```

```python
with pm.Model() as logistic_model:
    az.plot_forest(trace)
```

```python
with pm.Model() as logistic_model:
    varnames = ["educ"]
    az.plot_forest(trace, var_names=varnames);
```

```python
with pm.Model() as logistic_model:
    varnames = ["hours"]
    az.plot_forest(trace, var_names=varnames);
```

```python
b = trace['educ']
plt.hist(np.exp(b), bins=20, density=True)
plt.xlabel("Odds Ratio")
plt.show()
```

```python
lb, ub = np.percentile(b, 2.5), np.percentile(b, 97.5)
print("P(%.3f < O.R. < %.3f) = 0.95" % (np.exp(lb),np.exp(ub)))
```
