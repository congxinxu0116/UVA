```python
data7 = pandas.read_excel('C:\\David\\MYX\\Data\\Response_Variable.xlsx',
                          header = 0, 
                          skiprows=[1],
                          converters={'Scode': '{:0>6}'.format})
# data6 = data6[pandas.DatetimeIndex(data6['Accper']).month == 12]
data7 = data7[data7['Year'] != 2007]
data7.rename(columns={'Scode': 'Stkcd'}, inplace=True)
data7.head()
```

```python
final['Year'] = pandas.DatetimeIndex(final['Accper']).year
final = pandas.merge(data7, final, 
                     how='outer', 
                     on = ['Stkcd', 'Year'], 
                     indicator='Response_Variable')
final.head()
```

```python
print(min(final.Year))
print(max(final.Year))
print(len(final))
```

```python
import numpy
clean = final.copy()
clean['y'] = [1.0 if i == '标准无保留意见' else 0.0 for i in final['Audittyp']]
clean.head()

# Cannot add Stkcd for some reason...
formula = 'y ~ Year + WordNum + VocabularyNum + SentenceNum + \
            LM_positive + LM_negative + Ntusd_Positive + Ntusd_Negative + \
            LM_Tone1 + LM_Tone2 + Ntusd_Tone + B001100000  + \
            B001101000 + B001000000 + B002000000 + F010101A + \
            F010401A + F010601A + F010801B + F011201A + F012501B + \
            Meps + Mpe + Mnetpro + Mebit + BPS + ROA + ROE + \
            TotalAssetsTurnover + CompanySize + RegisterCapital'
```

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm

mod1 = smf.glm(formula=formula, data=clean, family=sm.families.Binomial()).fit()
print(mod1.summary())
```
