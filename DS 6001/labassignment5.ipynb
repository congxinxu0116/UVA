## Problem 1
Pull the HTML code from http://books.toscrape.com/. Make sure you provide a user agent string. Then parse this HTML code and save the parsed code as a separate Python variable. [3 points]

```python
# Store the URL as a variable
url = "http://books.toscrape.com/"
# Create a headers with user agent string
headers = {'user-agent': 'David Xu class assignment (cx2rx@virginia.edu)'}
# Get the URL page
r = requests.get(url, headers = headers)
# Parse the HTML with Beautiful Soup
books = BeautifulSoup(r.text, 'html.parser')
```

## Problem 2
Extract all 20 of the book titles and save them in a list. [2 points]

```python
BookTitle = list()
for i in range(0, 20):
    BookTitle += [books.find_all('article')[i].find_all('h3')[0].find_all('a')[0]['title']]

print(BookTitle)
```

## Problem 7
Create a function that takes the URL of the webpage to scrape as an input, applies the code you wrote for questions 1 through 6, and generates the dataframe from question 6 as the output. [3 points]

```python
def BookToScrape_Scraper(url): 
    # Create a headers with user agent string
    headers = {'user-agent': 'David Xu class assignment (cx2rx@virginia.edu)'}
    # Get the URL page
    r = requests.get(url, headers = headers)
    # Parse the HTML with Beautiful Soup
    books = BeautifulSoup(r.text, 'html.parser')
    
    # Scraping each books for title, price, star rating and cover image URL
    BookTitle = list()
    prices = list()
    starlevel = list()
    imageURL = list()
    
    for i in range(0, 20):
        BookTitle += [books.find_all('article')[i].find_all('h3')[0].find_all('a')[0]['title']]
        prices += [books.find_all('p', 'price_color')[i].string]
        prices = [s.replace('Â£', '') for s in prices]
        starlevel += [books.find_all('article')[i].find_all('p')[0]['class'][1]]
        imageURL += [url + books.find_all('article')[i].find_all('a')[0].find('img')['src']]
    
    # Create a data frame for return
    d = {'book titles': BookTitle, 'prices': prices, 'star ratings': starlevel, 'cover JPEG URLs': imageURL}
    df = pd.DataFrame(data=d)
    
    return(df)

BookToScrape_Scraper(url)
```
